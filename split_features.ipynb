```python
house = ['위치-집-정중앙',
             '문-격자무늬 창',
             '창문-창문을 가린 커튼',
             '위치-집-아래',
             '지붕-그물무늬 음영',
             '지면선-벽 밑면 있지만 지면 없음',
             '창문-많다',
             '창문-격자무늬'
             ]
tree = ['크기-나무-크다',
           '위치-나무-정중앙',
            '위치-나무-아래',
            '위치-나무-위',
            '절단-나무-위',
            '절단-나무-오른쪽',
            '가지-아래로 늘어진 가지',
            '뿌리-부재1',
            '수관-공 모양',
            '기둥-옹이구멍 안 동물',
            '가지-부재',
            '뿌리-부재2',
            '가지-옆으로 벌어짐',
            '기둥-상처, 옹이구멍'
            ]
person = ['크기-사람-작다',
              '위치-사람-정중앙',
              '위치-사람-오른쪽',
              '음영-사람-그림자',
              '음영-사람-내부의 진한 음영',
              '코-작다',
              '머리카락-윤곽선(검게 표현되지 않은 긴 머리)',
              '팔-길다',
              '손-부재',
              '손-크다',
              '손-원모양 손, 주먹쥔 손',
              '손-정교함(손톱, 주름)',
              '다리-넓게 벌림',
              '다리-앉아있음',
              '발-발가락 그림'
              ]
HTP = house + tree + person
len(HTP)
```

```python
sum = 0
for value in HTP:
    for num in range(1, data2.shape[1]):
        if data2[num]['레이블'] == value:
            print(value, num, data2[num].isnull().sum())
            sum = sum+1
print(sum, '개')
```

```python
data3 = pd.DataFrame()
for value in HTP:
    for num in range(1, data2.shape[1]):
        if data2[num]['레이블'] == value:
            data2[num].fillna(0, inplace=True)
            data3 = pd.concat([data3, data2[num]], axis=1)
            
data3.head()
```

```python
data3.columns = HTP
data3.head()
```

```python
data3 = data3.drop(['레이블'])
```
